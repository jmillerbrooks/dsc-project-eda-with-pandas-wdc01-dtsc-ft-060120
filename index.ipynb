```python
# Let's get started importing the necessary libraries
import pandas as pd
%matplotlib inline
```

```python
# Loading the data
df = pd.read_csv('ames_train.csv')
```

```python
# Investigate the Data
df.info()
```

```python
# Investigating Distributions using scatter_matrix
pd.plotting.scatter_matrix(df[['SalePrice', 'YearBuilt', 'YearRemodAdd', 'BedroomAbvGr', 'YrSold', 'SaleCondition' ]]);
```

```python
# Create a plot that shows the SalesPrice Distribution
df.SalePrice.plot.hist()
# df.SalePrice.plot.kde()
# df.SalePrice.plot.density()
# df.SalePrice.plot.box(showfliers=False)
```

```python
# Create a plot that shows the LotArea Distribution
# df.LotArea.plot.kde()
# df.LotArea.plot.box(showfliers=False)
df.loc[df['MSZoning'].str.startswith('R'), 'LotArea'].plot.kde()

```

```python
# Create a plot that shows the Distribution of the overall house condition
df.OverallCond.plot.kde()
```

```python
# Create a Box Plot for SalePrice
df.SalePrice.plot.box(showfliers=False)
```

```python
# Perform an Exploration of home values by age
df.plot.scatter('YearBuilt', 'SalePrice', c='YearRemodAdd', colormap='viridis');
```
